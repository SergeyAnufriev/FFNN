```python
%pylab inline

#Train a 2-layer network to fit the sine function in the interval [0.0, 7.0]

#Import the neural network module.
import ffnn

#Prepare the training data.
func = np.sin
net_input = np.array([np.linspace(0.0, 7.0)]).T
target = func(net_input)

#Define the structure of the network and parameters of the training.
structure = (1, 8, 1)
learning_rate = 0.1
max_epoch = 5000

#Create a network and train it.
net = ffnn.NN(structure, net_input, target)
net.train(learning_rate, max_epoch)

#Plot the output of the network together with and the sine function we attempt to fit.
net_output = net(net_input)
plot(net_input, target, 'r--', net_input, net_output, 'g-')
legend(('target', 'net_output'), loc=0)
grid(True)
title('Network output')
show()
```
